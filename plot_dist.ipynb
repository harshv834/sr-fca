```python
import torch
import os
import matplotlib.pyplot as plt
import yaml
```

```python
dataset="femnist"
seeds=[1729, 1739, 1749,1759, 1769]
dist_thresholds = {"femnist": 0.54 , "inv_mnist": 0.77}
```

```python
def plot_seed(dataset, seeds):
    all_data = []
    for seed in seeds:
        dist_dict_path = os.path.join("/base_vol/sr_fca_main/experiments", dataset , f"seed_{seed}","results/sr_fca/init/dist_dict.pth")
        dist_dict = torch.load(dist_dict_path)
        all_data += list(dist_dict.values())
    threshold = dist_thresholds[dataset]
    _ = plt.hist(all_data, bins=100, density=True, cumulative=False)
    threshold = sorted(dist_dict.values())[int(0.8* len(dist_dict))]
    plt.vlines(threshold, 0, 15, "r", linestyles="dashdot", label="lambda")
    plt.ylim(0,13)
    
    plt.ylabel("Normalized Frequency")
    plt.xlabel("Pairwise distance")
    plt.title(f"Normalized histogram of pairwise distances for {dataset}")
    plt.legend()
    
```

```python
plot_seed(dataset, seeds)
plt.savefig("femnist.jpg")
```
